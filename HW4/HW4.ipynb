```python
t = np.arange(-1000, 1000, 0.2)
x_s = []
for i in t:
  if np.floor((i + 0.5)) % 2 == 0:
    x_s.append(1)
  else:
    x_s.append(-1)
```

```python
def interpolate(x_s, t_vals, mid_i, T, N):
  x_c = []
  for t in t_vals:
    x_c_t = 0
    for n in range(mid_i - N, mid_i + N):
      x_c_t += x_s[n]*np.sinc((t - (n - mid_i)*T)/T)
    x_c.append(x_c_t)
  return x_c
```

```python
t_vals = np.arange(-20, 20, 0.01)
x_10 = interpolate(x_s, t_vals, 1000, 0.2, 10)
```

```python
plt.plot(t_vals, x_10)
plt.xlabel('t')
plt.ylabel('$x_{10}$')
```

```python
x_100 = interpolate(x_s, t_vals, 1000, 0.2, 100)
```

```python
plt.plot(t_vals, x_100)
plt.xlabel('t')
plt.ylabel('$x_{100}$')
```

```python
x_1000 = interpolate(x_s, t_vals, 1000, 0.2, 1000)
```

```python
plt.plot(t_vals, x_1000)
plt.xlabel('t')
plt.ylabel('$x_{1000}$')
```

As N increases, the reconstructed signal looks more like the continuous signal

2a)

```python
def const_a(w_c):
  return np.cos(np.pi/4 + w_c/2)/np.cos(np.pi/4 - w_c/2)
```

```python
def low_pass(z):
  return 0.25*(1 + 2*z**(-1) + z**(-2))
```

```python
def high_pass(z, w_c):
  a = const_a(w_c)
  return low_pass(-(z - a)/(1 - a*z))
```

```python
w = np.linspace(-np.pi, np.pi, 1000)
z = np.e **(1j * w)
```

```python
w_c = np.pi/4
H_hp = high_pass(z, w_c)
```

```python
plt.plot(w, abs(H_hp))
plt.title("2a: $w_{c} = \pi/4$")
plt.xlabel(r'$\omega$')
plt.ylabel(r'$|H_{hp}(e^{i \omega})|$')
```

```python
plt.plot(w, np.angle(H_hp))
plt.title("2a: $w_{c} = \pi/4$")
plt.xlabel(r'$\omega$')
plt.ylabel(r'$\phi(H_{hp}(e^{i \omega}))$')
```

```python
w_c = np.pi/2
H_hp = high_pass(z, w_c)
```

```python
plt.plot(w, abs(H_hp))
plt.title("2a: $w_{c} = \pi/2$")
plt.xlabel(r'$\omega$')
plt.ylabel(r'$|H_{hp}(e^{i \omega})|$')
```

```python
plt.plot(w, np.angle(H_hp))
plt.title("2a: $w_{c} = \pi/2$")
plt.xlabel(r'$\omega$')
plt.ylabel(r'$\phi(H_{hp}(e^{i \omega}))$')
```

```python
def all_pass(z, w_c):
  a = const_a(w_c)
  return -(z - a)/(1 - a*z)
```

```python
w_c = np.pi/4
H_ap = all_pass(z, w_c)
```

```python
plt.plot(w, abs(H_ap))
plt.title("2b: $w_{c} = \pi/4$")
plt.xlabel(r'$\omega$')
plt.ylabel(r'$|H_{ap}(e^{i \omega})|$')
```
